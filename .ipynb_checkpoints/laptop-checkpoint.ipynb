```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text =='Intel Core i5' or text =='Intel Core i3':
        return text
    else:
        if text.split()[0]=='Intel':
            return 'Other Intel Processor'
        else:
            return "AMD Processor"
```

```python
df['Cpu brand']=df['Cpu Name'].apply(fetch_processor)
```

```python

import re

# Convert GB/TB to numeric GB
def convert_to_gb(text):
    text = text.lower().strip()
    match = re.match(r'(\d+\.?\d*)\s*(GB|TB)', text)
    if match:
        value = float(match.group(1))
        unit = match.group(2)
        if unit == 'TB':
            value = value * 1024
        return int(value)
    return 0

def extract_storage_values(Memory):
     = hdd = flash = hybrid = 0
    parts = str(Memory).lower().split('+')

    for part in parts:
        part = part.strip()
        size_match = re.search(r'(\d+\.?\d*\s*(?:GB|TB))', part)
        size = convert_to_gb(size_match.group()) if size_match else 0

        if 'SSD' in part and 'nvme' not in part:
            SSD += size
        elif 'HDD' in part:
            HDD += size
        elif 'Flash' in part:
            Flash += size
        elif 'Hybrid' in part:
            Hybrid += size

    return pd.Series([SSD, HDD, Flash, Hybrid],index=['SSD', 'HDD', 'Flash_Storage', 'Hybrid'])

df[['SSD', 'HDD', 'Flash_Storage', 'Hybrid']] = df['Memory'].apply(extract_storage_values)

df.head()
```
